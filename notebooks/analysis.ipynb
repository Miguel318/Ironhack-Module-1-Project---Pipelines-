```python
#Import libraries:
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
plt.rcdefaults()
%matplotlib inline
```

```python
#Import clean data:
        #I used two datasets because joining the complete clean dataset with the scraping lost a lot of information

df1 = pd.read_csv('/Users/MIGUEL/Desktop/CLAB1/Ironhack-Module-1-Project---Pipelines-/data/processed/df_clean.csv')
df2 = pd.read_csv('/Users/MIGUEL/Desktop/CLAB1/Ironhack-Module-1-Project---Pipelines-/data/processed/df_project2CLEAN.csv')

display(df1)
display(df2)
```

```python
#Graphic_1: Wealth distribution
graphic_1 = df1['worth BUSD'].hist(bins=100, figsize=(16,8))
plt.xlabel('Position')
plt.ylabel('Wealth')
plt.title("Wealth distribution")
plt.savefig("Wealth distribution")
```

```python
#Graphic_2: Scattter - Wealth distribution
Graphic_2 = df1.plot("position", "worth BUSD", kind="scatter")
plt.savefig("Scattter - Wealth distribution") 
```

```python
#Graphic_3: Scatter worthChange 
graphic_3 = df1.plot("position", " worthChange millions USD", kind="scatter")
plt.savefig("scatter position-worthchange")
```

```python
#Graphic_4: Wealth comparison between genders
genders_wealth_comparison = df1.groupby('gender')['worth BUSD'].sum().sort_values(ascending=False)/100
gender = ['Male', 'Female', 'None']
colors = ['skyblue', 'grey', 'white']
genders_wealth_comparison.plot(kind='pie', labels = gender, colors = colors, startangle=45, shadow=False, figsize=(16,8),  autopct='%.2f%%')
plt.title('Wealth comparison between genders')
plt.savefig("Wealth comparison between genders")

```

```python
#Graphic_5: Wealth comparison between countries 
countries_wealth_comparison = df2.groupby(['country', 'worth BUSD']).agg(['count', 'mean'])
fig = plt.figure(figsize=(50,50))
df1.country[df2.position].value_counts(normalize = True).plot(kind='barh', alpha=0.5)
plt.title('Wealth comparison between countries')
plt.savefig("Wealth comparison between countries") 
```

```python
#Graphic_6: Number of billionaires per country

colors = ['grey', 'mediumvioletred', 'gold', 'orangered', 'yellowgreen', 'r', 'skyblue', 'navy', 'grey', 'black']
df2['country'].value_counts()[:15].plot(kind='bar', width=0.9, figsize=(20,10), color = colors)
plt.xlabel('Country (Top 30)')
plt.ylabel('Number of billionaires')
plt.title('Number of billionaires per country')
plt.savefig("Number of billionaires per country") 
```
